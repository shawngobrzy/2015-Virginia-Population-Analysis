```python
#Read the data into a dataframe

import pandas as pd

census_df = pd.read_csv('acs2015_county_data.csv')
```

```python
#Display data

census_df.head()
```

```python
#List the columns being used in the dataset

census_df.info()
```

```python
#Describe the data

census_df.describe(include='all')
```

```python
#Filter state data down to that of only Virginia

import numpy as np

specific_state = 'Virginia'

criteria = census_df["State"] == specific_state

virginia_df = census_df[criteria]

virginia_df.head()
```

```python
#Check for missing values

virginia_df.isnull().sum()
```

```python
virginia_df.columns = virginia_df.columns.str.lower()
virginia_df.columns = virginia_df.columns.str.replace(' ', '_')
virginia_df = virginia_df.rename(columns={'censusid':'census_id'})
virginia_df = virginia_df.rename(columns={'incomerr':'income_err'})
virginia_df = virginia_df.rename(columns={'incomepercap':'income_per_cap'})
virginia_df = virginia_df.rename(columns={'incomepercaperr':'income_per_cap_err'})
virginia_df = virginia_df.rename(columns={'childpoverty':'child_poverty'})
virginia_df = virginia_df.rename(columns={'othertransp':'other_transp'})
virginia_df = virginia_df.rename(columns={'workathome':'work_at_home'})
virginia_df = virginia_df.rename(columns={'meancommute':'mean_commute'})
virginia_df = virginia_df.rename(columns={'privatework':'private_work'})
virginia_df = virginia_df.rename(columns={'publicwork':'public_work'})
virginia_df = virginia_df.rename(columns={'selfemployed':'self_employed'})
virginia_df = virginia_df.rename(columns={'familywork':'family_work'})

virginia_df.columns
```

```python
# Melt the genders columns into a single column
virginia_gender_df = virginia_df.melt(
    id_vars=["census_id", "state", "county", "totalpop", "hispanic",
       "white", "black", "native", "asian", "pacific", "citizen", "income",
       "incomeerr", "income_per_cap", "income_per_cap_err", "poverty",
       "child_poverty", "professional", "service", "office", "construction",
       "production", "drive", "carpool", "transit", "walk", "other_transp",
       "work_at_home", "mean_commute", "employed", "private_work",
       "public_work", "self_employed", "family_work", "unemployment"],   # columns to keep
    value_vars=["men","women"],  # columns to unpivot
    var_name="gender",     # new column name for gender
    value_name="genderpop"  # new column name for population
)

virginia_gender_df.head()
```

```python
# Melt the worktype columns into a single column
virginia_worktype_df = virginia_df.melt(
    id_vars=["census_id", "state", "county", "totalpop", "men", "women", "hispanic",
       "white", "black", "native", "asian", "pacific", "citizen", "income",
       "incomeerr", "income_per_cap", "income_per_cap_err", "poverty",
       "child_poverty", "drive", "carpool", "transit", "walk", "other_transp",
       "work_at_home", "mean_commute", "employed", "private_work",
       "public_work", "self_employed", "family_work", "unemployment"],   # columns to keep
    value_vars=["professional", "service", "office", "construction",
       "production"],  # columns to unpivot
    var_name="worktype",     # new column name for worktype
    value_name="worktypepop"  # new column name for population
)

virginia_worktype_df.head()
```

```python
# Melt the workpoint columns into a single column
virginia_workpoint_df = virginia_df.melt(
    id_vars=["census_id", "state", "county", "totalpop", "men", "women", "hispanic",
       "white", "black", "native", "asian", "pacific", "citizen", "income",
       "incomeerr", "income_per_cap", "income_per_cap_err", "poverty",
       "child_poverty", "professional", "service", "office", "construction",
       "production", "drive", "carpool", "transit", "walk", "other_transp",
       "work_at_home", "mean_commute", "employed", "unemployment"],   # columns to keep
    value_vars=["private_work",
       "public_work", "self_employed", "family_work"],  # columns to unpivot
    var_name="workpoint",     # new column name for workpoint
    value_name="workpointpop"  # new column name for population
)

virginia_worktype_df.head()
```

```python
# Melt the race columns into a single column
virginia_race_df = virginia_df.melt(
    id_vars=["census_id", "state", "county", "totalpop", "men", "women", "citizen", "income",
       "incomeerr", "income_per_cap", "income_per_cap_err", "poverty",
       "child_poverty", "professional", "service", "office", "construction",
       "production", "drive", "carpool", "transit", "walk", "other_transp",
       "work_at_home", "mean_commute", "employed", "private_work",
       "public_work", "self_employed", "family_work", "unemployment"],   # columns to keep
    value_vars=["hispanic","white", "black", "native", "asian", "pacific"],  # columns to unpivot
    var_name="race",     # new column name for race
    value_name="racepop"  # new column name for population
)

virginia_race_df.head()
```
